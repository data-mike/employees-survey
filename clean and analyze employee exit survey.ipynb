# Identify Dissatisfied Employees

We will identify any employees who resigned because they were dissatisfied. Below are columns we'll use to categorize employees as "dissatisfied" from each dataframe:

1. tafe_survey_updated:
  
    * Contributing Factors. Dissatisfaction
    * Contributing Factors. Job Dissatisfaction
    
2. dete_survey_updated:
    
    * job_dissatisfaction
    * dissatisfaction_with_the_department
    * physical_work_environment
    * lack_of_recognition
    * lack_of_job_security
    * work_location
    * employment_conditions
    * work_life_balance
    * workload
    


If any of the employees indicated any of the factors above caused them rot resign, we'll mark them as *dissatisfied* in a new column.

To create the new column, we'll do the following :

1. Convert the values in the 'Contributing Factors. Dissatisfaction' and 'Contributing Factors. Job Dissatisfaction' columns in the *tafe_resignations* dataframe to True, False, or NaN values.

2. We will use the DataFrame.any() method to add a *True* value to the *dissatisfied* column if any of the columns listed above contain a True value.

```python
# we will explore the 'Contributing Factors. Dissatisfaction' and 
# 'Contributing Factors. Job Dissatisfaction' of the tafe_resignations

tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts(
    dropna=False)
```

```python
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts(
dropna=False)
```

```python
# we update the values in these two columns of tafe_resignations so that each
# contains only True, False, or Nan values:

def update_vals(val_):
    if pd.isnull(val_):
        return np.nan
    elif val_== '-':
        return False
    else:
        return True
```

```python
# use the df.applymap() to apply the above function

tafe_resignations['dissatisfied'] = tafe_resignations[
    ['Contributing Factors. Dissatisfaction',
                  'Contributing Factors. Job Dissatisfaction']].applymap(
                      update_vals).any(axis=1, skipna=False)

tafe_resignations_up = tafe_resignations.copy()

tafe_resignations_up.dissatisfied.value_counts()


```

```python
dete_resignations['dissatisfied'] = dete_resignations[
    ['job_dissatisfaction','dissatisfaction_with_the_department',
     'physical_work_environment','lack_of_recognition','lack_of_job_security',
     'work_location','employment_conditions','work_life_balance','workload']
].any(axis=1, skipna=False)

dete_resignations_up = dete_resignations.copy()

dete_resignations_up['dissatisfied'].value_counts(dropna=False)
```

# Combine the Data

We will combine both datasets and add an 'institue' column to easily distinguish between the two data

We still have some columns left in the dataframe that we don't need to complete our analysis. We will use the df.dropna() method to drop any columns with less than 500 non-null values.

```python
# Removing columns with less than 500 non-null values using df.dropna()

combined_updated = combined.dropna(thresh=500, axis=1).copy()

combined_updated.info()  # check that the right columns were indeed dropped
```

# Clean the Service Column

We will clean the *institute_service* column and categorize it according to the definitions below:

 * New: less than 3 years at a company
 * Experienced: 3-6 years at a company
 * Established: 7-10 years at a company
 * Veteran: 11 or more years at a company

```python
# explore the unique values in the institute_service column

combined_updated.institute_service.value_counts(dropna=False)
```

```python
# convert the institute_service column to 'str' using the astype() method

combined_updated['institute_service'] = combined_updated[
    'institute_service'].astype('str')

```

```python
# extract the years using the series.str.extract() method

combined_updated['years_extract'] = combined_updated[
    'institute_service'].str.extract(r'(\d+)').astype('float')

combined_updated.years_extract.value_counts(dropna=False).sort_index()
```

```python
# we'll create a function that maps each year value to the defined category we
# want

def categorize_(num):
    if num >= 11:
        return 'Veteran'
    elif 7<= num < 11:
        return 'Established'
    elif 3 <= num < 7:
        return 'Experienced'
    elif pd.isnull(num):
        return np.nan
    else:
        return 'New'
```

```python
combined_updated['service_cat'] = combined_updated['years_extract'
                                                  ].apply(categorize_)

combined_updated.service_cat.value_counts(dropna=False)
```

# Perform Initial Analysis

Finally, we are going to replace the missing values in the *dissatisfied* column by replacing with the most frequent value - False. Afterwards, we will calculate the percentage of employees who resigned due to dissatisfaction for each category in the *service_cat* column and we'll plot the result.

It should be noted that we still have additional missing values left to deal with. This is meant to be an intial introduction to the analysis, not the final analysis.

```python
# explore the 'dissatisfied' column

combined_updated.dissatisfied.value_counts(dropna=False)
```

```python
combined_updated['dissatisfied'] = combined_updated['dissatisfied'
                                                    ].fillna(False)

```

```python
dissat_pct = combined_updated.pivot_table(
    index=['service_cat'], values=['dissatisfied'],
                            aggfunc=np.mean)

#plot the result from the pivot_table above

%matplotlib inline

dissat_pct.plot(kind='bar', rot=30)
```

From the plot above, we can conclude that employees with more than 7 years of experience are more likely to resign due to dissatisfaction than employees with less than 7 years of experience. 

However, resolving the missing values will provide a more accurate picture of employees resignation analysis.
